## 3 - Velocity

We compute an estimate for the velocity using change in position over time. Since the data is split into (x,y) components already, it makes sense to comute the velocity component wise also.

```python
def calculate_player_velocity(team,max_speed=12):
    
    # Get player id's used in team data frame
    player_ids = np.unique([c[:-2] for c in tracking_home.columns if c[:4] in ['Home','Away']])
    
    # Get vector of time steps
    dt = tracking_home["Time [s]"].diff()
    
    for id in player_ids:
        # Calculate raw velocity components
        vx = team[id+"_x"].diff() / dt
        vy = team[id+"_y"].diff() / dt
        
        
        if max_speed > 0:
            speed_sq = (vx**2+vy**2)
            
            # Remove instances when player speed is above max, this is likely due to data errors or eg half time
            vx[speed_sq>max_speed**2] = np.nan
            vy[speed_sq>max_speed**2] = np.nan
        
        team[id+"_vx"] = vx
        team[id+"_vy"] = vy
    return team
```

This looks quite 'jagged'. The above graph is over a 2 min period. We could smooth the data out.

```python
def remove_player_velocities(team):
    # remove player velocoties and acceleeration measures that are already in the 'team' dataframe
    columns = [c for c in team.columns if c.split('_')[-1] in ['vx','vy','ax','ay','speed','acceleration']] # Get the player ids
    team = team.drop(columns=columns)
    return team
```

```python
def savgol_filter(x,window,polyorder):
    if(len(x)>window):
        #x = x[~np.isnan(x)] # Drop nan values
        return sci.signal.savgol_filter(x,window,polyorder)
    else:
        return x
```

```python
def calculate_player_velocity(team,max_speed=12,smoothing=True,filter_='Savitzky-Golay',window=7,polyorder=1):
    team = remove_player_velocities(team)
    
    # Get player id's used in team data frame
    player_ids = np.unique([c[:-2] for c in team.columns if c[:4] in ['Home','Away']])
    
    # index of first frame in second half
    second_half_idx = team.Period.idxmax(2)
    
    # Get vector of time steps
    dt = tracking_home["Time [s]"].diff()
    
    for player in player_ids:

        # Calculate raw velocity components
        vx = team[player + "_x"].diff() / dt
        vy = team[player + "_y"].diff() / dt
        
        vx.fillna(0,inplace=True)
        vy.fillna(0,inplace=True)
        
        # Remove instances when player speed is above max, this is likely due to data errors or eg half time
        if max_speed > 0:
            speed_sq = (vx**2+vy**2)
            vx[speed_sq>max_speed**2] = np.nan
            vy[speed_sq>max_speed**2] = np.nan
        
        # Interpolates nans
        vx.interpolate(inplace=True)
        vy.interpolate(inplace=True)
            
        if smoothing:
          
            if (filter_=='Savitzky-Golay'):
                vx[:second_half_idx] = savgol_filter(vx[:second_half_idx],window,polyorder)
                vy[:second_half_idx] = savgol_filter(vy[:second_half_idx],window,polyorder)
                vx[second_half_idx:] = savgol_filter(vx[second_half_idx:],window,polyorder)
                vy[second_half_idx:] = savgol_filter(vy[second_half_idx:],window,polyorder)
        
        epsilon = 0.0001
        vx[abs(vx)<epsilon] = 0
        vy[abs(vy)<epsilon] = 0
                
        team[player+"_vx"] = vx
        team[player+"_vy"] = vy
    return team
```

## 4 - Ball Arrival Times

Now we have smooth player velocity we can look at the time it takes for a football to arrive

```python
m = 0.45 # In kg
g = -9.81
vt = -35 # Terminal velocity downwards
k = m*g / vt # Drag coefficient
```

```python
def T(v):
    q = 1 - ((k*v) / (m*g)) # Just a constant
    
    return (m/k - v/g + (m/k * sci.special.lambertw(-q*np.exp(-q)))).real
```

```python
def s_y(t,v_y):
    return (m*g*t)/k + (m/k)*(v_y-(m*g/k))*(1-np.exp(-k*t/m))
```

```python
def s_x(t,v_x):
    return m*v_x/k * (1-np.exp(-k*t/m))
```

```python
v_x = 44.7 / np.sqrt(2)
v_y = 44.7 / np.sqrt(2)
```

```python
ranges = pd.DataFrame(fun_map)
ranges.columns = v_x_l
ranges.index = v_y_l
```

```python
def time_interval(d):
    df = abs(ranges-d)
    df=df[df<0.2]
    df = df[df.columns[~df.isnull().all()]]
    h = df.columns.values
    
    return [T(h[0]),T(h[-1])] if len(h)>0 else [np.nan,np.nan]
```

```python
time_ranges = [[d,time_interval(d)] for d in np.linspace(0,65) if not np.isnan(time_interval(d)).any()]
```

## 5 - Player Arrival Times

```python
def player_arrival_times(tracking_home_slice,tracking_away_slice,ball_x,ball_y,vel_mult_factor = 3,avg_running_speed = 5,reaction_time=0.3):
    
    home_arrival_times = pd.Series([],dtype=pd.StringDtype()) 
    away_arrival_times = pd.Series([],dtype=pd.StringDtype()) 
    
    for team,arrival_times in zip([tracking_home_slice,tracking_away_slice],[home_arrival_times,away_arrival_times]) :
        # Get the player ids
        player_ids = np.unique( [ c[:-2] for c in team.keys() if (c[:4] in ['Home','Away'] and c[-2]!='v') ] )
        for player in player_ids:
            pos_and_speed = (team[player+"_x"],team[player+"_y"],team[player+"_vx"],team[player+"_vy"])
            
            # If all position and speed is well defined
            if(not any(np.isnan(pos_and_speed))):
                x,y,v_x,v_y = pos_and_speed
                dist_to_ball =  np.linalg.norm(np.abs([x-ball_x,y-ball_y]))
                
                u = [x+v_x,y+v_y] # Vector in direction of current motion
                v = [ball_x-x,ball_y-y] # Pos vector of the ball from player pos
                
                cos_alpha = np.dot(u,v)/np.linalg.norm(u)/np.linalg.norm(v)
                alpha = np.abs(np.arccos(np.clip(cos_alpha, -1, 1)))
                
                if dist_to_ball < 1:
                    arrival_times[player+"_arrival_time"] = 0
                elif dist_to_ball < 5 and alpha > 3.14/2: # If close & bezier would be too extreme
                    dist_to_go =  np.linalg.norm([x+reaction_time*v_x-ball_x,y+reaction_time*v_y-ball_y])
                    arrival_times[player+"_arrival_time"] = reaction_time + dist_to_go / avg_running_speed
                else: # Bezier curve approach
                    nodes = np.asfortranarray([
                        [x+reaction_time*v_x,x+vel_mult_factor*v_x,ball_x],
                        [y+reaction_time*v_y,y+vel_mult_factor*v_y,ball_y]
                    ])

                    path = bezier.Curve(nodes,degree=2)
                    arrival_times[player+"_arrival_time"] = path.length / avg_running_speed
            else:
                arrival_times[player+"_arrival_time"] = np.nan
                
    return home_arrival_times,away_arrival_times
                
```

## 6 - Starting To Put This All Together

```python
def default_model_params(time_to_control_veto=3):
    # key parameters for the model, as described in Spearman 2018
    params = {}
    # model parameters
    params['tti_sigma'] = 0.45 # Standard deviation of sigmoid function in Spearman 2018 ('s') that determines uncertainty in player arrival time
    params['lambda'] = 4.3 # ball control parameter
    params['lambda_gk'] = params['lambda']*3.0 # make goal keepers must quicker to control ball (because they can catch it)
    
    # numerical parameters for model evaluation
    params['int_dt'] = 0.04 # integration timestep (dt)
    params['max_int_time'] = 10 # upper limit on integral time
    params['model_converge_tol'] = 0.05 # assume convergence when PPCF>0.99 at a given location.
    
    # The following are 'short-cut' parameters. We do not need to calculated PPCF explicitly when a player has a sufficient head start. 
    # A sufficient head start is when the a player arrives at the target location at least 'time_to_control' seconds before the next player
    params['time_to_control'] = time_to_control_veto*np.log(10) * (np.sqrt(3)*params['tti_sigma']/np.pi + 1/params['lambda'])
    return params
```

```python
params = default_model_params()
```

```python
def probability_intercept_ball(time_to_intercept,T,params):
    # probability of a player arriving at target location at time 'T' given their expected time_to_intercept (time of arrival), as described in Spearman 2018
    f = 1/(1. + np.exp( -np.pi/np.sqrt(3.0)/params['tti_sigma'] * (T-time_to_intercept) ) )
    return f
```

```python
def calculate_pitch_control_at_target(target,tracking_home,tracking_away,time,params):
    # Set Up
    ball_start = [tracking_home.iloc[time]["ball_x"],tracking_home.iloc[time]["ball_y"]]
    
    ## 1 - Ball Travel Time Range
    dist = np.linalg.norm(np.asarray(target)-np.asarray(ball_start)) # Dist ball has to travel
    if dist < time_ranges[-1][0]:
        ball_flight_time_interval = time_ranges[np.argmin([np.abs(t[0]-dist) for t in time_ranges])][1] # range of times the ball could take to get there
    else:
        return np.nan , np.nan # No team can control this point
    
    ## 2 - Player Arrival Times
    home_arrival_times,away_arrival_times = player_arrival_times(tracking_home.iloc[time],tracking_away.iloc[time],target[0],target[1])
    # Remove nans and sort quickest to slowest
    home_arrival_times = np.sort(home_arrival_times[~np.isnan(home_arrival_times)])
    away_arrival_times = np.sort(away_arrival_times[~np.isnan(away_arrival_times)])
    
    # Set up player arrays
    home_players = [{"arrival_time":x,"PPCF":0} for x in home_arrival_times]
    away_players = [{"arrival_time":x,"PPCF":0} for x in away_arrival_times]
    
    tau_min_home = home_arrival_times[0]
    tau_min_away = away_arrival_times[1]
    
    ## 3 - Ball travel time
    ## For the moment just take a normal flight time, could skew this to favour one team or the other
    skew = 0.5 # Can weight to make ball arrive earlier or later, or pick this based on th team with the ball
    ball_travel_time = skew*ball_flight_time_interval[1]+(1-skew)*ball_flight_time_interval[1]
    
    ## 4 - Solving eqtn 3 in the paper
    # Check whether we actually need to solve equation 3
    if tau_min_home-max(ball_travel_time,tau_min_away) >= params['time_to_control']:
        # if away team can arrive significantly before homw team, no need to solve pitch control model
        return 0., 1.
    elif tau_min_away-max(ball_travel_time,tau_min_home) >= params['time_to_control']:
        # if home team can arrive significantly before away team, no need to solve pitch control model
        return 1., 0.
    else: 
        # solve pitch control model by integrating equation 3 in Spearman et al. ]
        # set up integration arrays
        dT_array = np.arange(ball_travel_time-params['int_dt'],ball_travel_time+params['max_int_time'],params['int_dt']) 
        PPCFhome = np.zeros_like( dT_array )
        PPCFaway= np.zeros_like( dT_array )
        # integration equation 3 of Spearman 2018 until convergence or tolerance limit hit (see 'params')
        ptot = 0.0
        i = 1
       
        while 1-ptot>params['model_converge_tol'] and i<dT_array.size: 
            T = dT_array[i]
            for player in home_players:
                # calculate ball control probablity for 'player' in time interval T+dt
                dPPCFdT = (1-PPCFhome[i-1]-PPCFaway[i-1])*probability_intercept_ball(player['arrival_time'],T,params) * params['lambda']
                # make sure it's greater than zero
                assert dPPCFdT>=0, 'Invalid homeacking player probability (calculate_pitch_control_at_target)'
                player['PPCF'] += dPPCFdT*params['int_dt'] # total contribution from individual player
                PPCFhome[i] += player['PPCF'] # add to sum over players in the homeacking team (remembering array element is zero at the start of each integration iteration)
            for player in away_players:
                # calculate ball control probablity for 'player' in time interval T+dt
                dPPCFdT = (1-PPCFhome[i-1]-PPCFaway[i-1])*probability_intercept_ball(player['arrival_time'],T,params) * params['lambda']
               # make sure it's greater than zero
                assert dPPCFdT>=0, 'Invalid awayending player probability (calculate_pitch_control_at_target)'
                player['PPCF'] += dPPCFdT*params['int_dt'] # total contribution from individual player
                PPCFaway[i] += player['PPCF'] # add to sum over players in the awayending team
            ptot = PPCFaway[i]+PPCFhome[i] # total pitch control probability 
            i += 1
        if i>=dT_array.size:
            print("Integration failed to converge: %1.3f" % (ptot) )
        return PPCFhome[i-1], PPCFaway[i-1]
```

Now let's iterate this over the whole pitch

```python
def generate_pitch_control(tracking_home, tracking_away,time,params, field_dimen = (106.,68.,), n_grid_cells_x = 50):

    # break the pitch down into a grid
    n_grid_cells_y = int(n_grid_cells_x*field_dimen[1]/field_dimen[0])
    dx = field_dimen[0]/n_grid_cells_x
    dy = field_dimen[1]/n_grid_cells_y
    xgrid = np.arange(n_grid_cells_x)*dx - field_dimen[0]/2. + dx/2.
    ygrid = np.arange(n_grid_cells_y)*dy - field_dimen[1]/2. + dy/2.
    
    # initialise pitch control grids for home and away teams
    PPCFhome = np.zeros( shape = (len(ygrid), len(xgrid)) )
    PPCFaway = np.zeros( shape = (len(ygrid), len(xgrid)) )


    # calculate pitch pitch control model at each location on the pitch
    checksum = 0
    num_grid_squares_in_range = 0
    for i in range( len(ygrid) ):
        for j in range( len(xgrid) ):
            target = np.array( [xgrid[j], ygrid[i]] )
            PPCFhome[i,j],PPCFaway[i,j] = calculate_pitch_control_at_target(target,tracking_home,tracking_away,time,params)
            
            if np.isfinite(PPCFhome[i,j]) and np.isfinite(PPCFaway[i,j]):
                checksum += PPCFhome[i,j]+PPCFaway[i,j]
                num_grid_squares_in_range +=1
    
    assert checksum/num_grid_squares_in_range > 0.95
    print(checksum/num_grid_squares_in_range)
    return PPCFhome,PPCFaway
```

```python
def plot_pitchcontrol_at_time(tracking_home, tracking_away, time, params,n_grid_cells_x = 100, alpha = 0.7, include_player_velocities=True, annotate=False, field_dimen = (106.0,68)):    
    # Calculate pitch control surface
    PPCFhome,PPCFaway = generate_pitch_control(tracking_home, tracking_away,time,params, field_dimen, n_grid_cells_x)
    # plot frame
    fig,ax = mviz.plot_pitch(field_color='white', field_dimen = field_dimen)
    mviz.plot_frame(tracking_home.loc[time], tracking_away.loc[time], figax=(fig,ax), PlayerAlpha=alpha, include_player_velocities=include_player_velocities, annotate=annotate )
    
    # plot pitch control surfaceT
    if True=='Home':
        PPCF = PPCFhome
        cmap = 'bwr'
    else:
        PPCF = PPCFaway
        cmap = 'bwr_r'
    ax.imshow(np.flipud(PPCF), extent=(-field_dimen[0]/2., field_dimen[0]/2., -field_dimen[1]/2., field_dimen[1]/2.),interpolation='spline36',vmin=0.0,vmax=1.0,cmap=cmap,alpha=0.5)

    return fig,ax
```

```python
time = 1000
fig,ax = plot_pitchcontrol_at_time(tracking_home, tracking_away, time, params,n_grid_cells_x = 10, alpha = 0.7, include_player_velocities=True, annotate=False, field_dimen = (106.0,68))
```

```python
def save_pitch_control_frame_at_time(T):
    fig,ax = plot_pitchcontrol_at_time(tracking_home, tracking_away,T, params,n_grid_cells_x = 50, alpha = 0.7, include_player_velocities=True)
    fig.savefig("K:\Code\Football Analysis\Pitch Control Images\{T}.png".format(T=T),bbox_inches='tight', pad_inches=0.05)
    plt.close()
    return T
```

```python
start_time = 2596
duration = 10
fps = 25

start_frame = start_time * fps

dt = np.arange(start_frame,start_frame+duration*fps,1)
delayed_results = [delayed(save_pitch_control_frame_at_time)(T) for T in dt]
```
